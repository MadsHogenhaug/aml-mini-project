```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
img_dim = 28
cats = np.load('../data/cats/full_numpy_bitmap_cat.npy').reshape(-1, 1, img_dim, img_dim) / 255
```

```python
# plot grid of cats
n_cats = 10
fig, ax = plt.subplots(n_cats, n_cats, figsize=(10, 10))
for i in range(n_cats):
    for j in range(n_cats):
        ax[i, j].imshow(cats[i*5 + j][0], cmap='gray')
        ax[i, j].axis('off')
plt.tight_layout()
```
